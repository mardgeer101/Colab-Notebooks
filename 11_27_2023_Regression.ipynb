```python
# Creating bigger data sets
X = tf.range(-100,100,4)
X
```

```python
y = X + 10
y
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X,y);
```

# The Three sets
1. the training set - model learns from this data, typically 70% to 80% of the total data
2. the validation set - model gets tuned in this data, typically 10% to 20% of the total data
3. the testing set - model gets evaaluated on this data what it has learned, typically 10% to 20% of the total data

```python
#spliting the data set
len(X), len(y)
```

```python
X_train = X[:40] # 80% of the total data
y_train = y[:40]

X_test = X[40:] # 20# of the total data
y_test = y[40:]

X_train, X_test, y_train, y_test
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X_train,y_train,c="b",label = "Training data")
plt.scatter(X_test,y_test,c="g",label = "Testing data")
plt.legend()
```

```python
#create model
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10,input_shape=[1],name="Input_layer"),
    tf.keras.layers.Dense(50,input_shape=[1],name="hidden_layer_1"),
    tf.keras.layers.Dense(10,input_shape=[1],name="hidden_layer_2"),
    tf.keras.layers.Dense(1,name="output_layer")
],name="model_1")

#compile the model
model.compile(loss = tf.keras.losses.mse,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics = (["mse"]))
```

```python
model.summary()
```

```python
#fit the model
model.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model=model,show_shapes=True)
```

```python
# make some predictions
y_pred = model.predict(X_test)
y_pred, y_test
```

```python
# creating a ploting function

def plot_predictions(train_data = X_train,
                     train_label = y_train,
                     test_data = X_test,
                     test_label = y_test,
                     predictions = y_pred):
  plt.figure(figsize=(10,7))
  plt.scatter(train_data,train_label, c="b", label="Training data")
  plt.scatter(test_data,test_label, c="g", label ="Test data")
  plt.scatter(test_data,predictions,c="r", label = "Predictions")
  plt.legend()

plot_predictions()
```

```python
model.evaluate(X_test, y_test)
```

```python
def mae(y_test,y_pred):
  return tf.metrics.mean_absolute_error(y_true=y_test,
                                        y_pred=y_pred)
def mse(y_test,y_pred):
  return tf.metrics.mean_squared_error(y_true=y_test,
                                      y_pred=y_pred)
```
