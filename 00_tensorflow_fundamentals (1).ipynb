# Importing tensorflow libraries


```python
import tensorflow as tf
print(tf.__version__)
```

#Creating tensors using `tf.constant()`
* scalar
* vector
* matrix
* tensor

```python
scalar = tf.constant(5,dtype=(tf.float32))
print("scalar information:",scalar)
print("scalar shape:",scalar.shape)
print("scalar data type:",scalar.dtype)
print("scalar valaue:",scalar.numpy())
print("scalar dimension:",scalar.ndim)
```

```python
vector = tf.constant([4,5],dtype=(tf.float16))
print("vector information:",vector)
print("vector shape:",vector.shape)
print("vector data type:",vector.dtype)
print("vector value:",vector.numpy())
print("vector dimension",vector.ndim)
```

```python
matrix = tf.constant([[2,4],
                      [5,8],
                      [1,7]],dtype=(tf.float32))
print("matrix information:\n",matrix)
print("matrix shape:",matrix.shape)
print("matrix data type:",matrix.dtype)
print("matrix value:\n",matrix.numpy())
print("matrix dimension:", matrix.ndim)
```

```python
tensor = tf.constant([[[3,4,6],[7,5,8],[2,4,6]],
                      [[7,3,6],[2,8,2],[5,8,9]]],dtype=tf.float32)
print("tensor information:",tensor)
print("tensor shape:",tensor.shape)
print("tensor data type:",tensor.dtype)
print("tensor value:\n",tensor.numpy())
print("tensor dimension:",tensor.ndim)
```

# Shuffle a tensor using `tf.random.shuffle`

```python
tf.random.set_seed(42)
x = tf.random.uniform([3,3],dtype=tf.int32,minval=0,maxval=50)
y = tf.random.uniform([2,3],dtype=tf.int32,minval=0,maxval=50)
x,y
```

```python
x_shuffle = tf.random.shuffle(x)
y_shuffle = tf.random.shuffle(y)
x_shuffle ,y_shuffle
```

# Multiplication of tensors using:
* `tf.reshape` and `tf.transpose`
  1. `*`
  2. `tf.matmul`
  3. `@`
  4. `tf.tensordot`

```python
tf.random.set_seed(42)
w = tf.random.uniform([3,3],dtype=tf.int32, minval=0,maxval=100,seed=10)
x, y, x_reshape, y_reshape, x_transpose, y_transpose
```

```python
print(x*w)
print(tf.matmul(x,w))
print(x@w)
print(tf.tensordot(x,w,1))
```

```python
print(tf.matmul(x,y_reshape))
print(tf.tensordot(x,y_reshape,1))
print(x@y_reshape)
print("\n")
print(tf.matmul(x,y_transpose))
print(tf.tensordot(x,y_transpose,1))
print(x@y_transpose)
```

# Finding the max, min, mean, sum, variance and std_dev of a tensor using:
* `tf.reduce_max`
* `tf.reduce_min`
* `tf.reduce_mean`
* `tf.reduce_sum`
* `tf.math.reduce_variance`
* `tf.math.reduce_std`

```python
tf.random.set_seed(42)
sample_tensor = tf.random.uniform([2,3,3],maxval=100,minval=0)
sample_tensor
```

```python
print(tf.reduce_max(sample_tensor))
print(tf.reduce_min(sample_tensor))
print(tf.reduce_mean(sample_tensor))
print(tf.reduce_sum(sample_tensor))
print(tf.math.reduce_variance(sample_tensor))
print(tf.math.reduce_std(sample_tensor))
```

# Finding the positional maximum and minimum of a tensor using:
* `tf.math.argmax`

```python
tf.random.set_seed(42)
another_sample_tensor = tf.random.uniform([2,2,5],dtype=tf.int32,maxval = 100, minval=0,seed=10)
another_sample_tensor
```

```python
pmax = tf.math.argmax(another_sample_tensor,1)
pmax
```

```python
pmin = tf.math.argmin(another_sample_tensor,1)
pmin
```
