## steps in modelling with tensorflow
1. **creating a model** - define the input and output layers, as well as the hidden layers of a deep learning model.
2. **compiling a model** - define the loss function (in other words, the function which tells our model how wrong it is) and the optimizer (tells our model how to improve the patterns its learning) and evaluation metrics (what we can use to interpret the performance of our model).
3. **fitting a model** - letting a model try to find patterns between x and y or features and labels

## Common ways to improve a deep model
  * **Adding Layers** ---> `tf.keras.Dense()`
  * **Increase the number of hidden units or neurons** ---> `tf.keras.Dense(100)`
  * **Change the activation functions** ---> `tf.keras.Dense(100, activation="relu")`
  * **Change the optimization function** --->`optimizer = tf.keras.optimizers.Adam()`
  * **change the learning rate or** `learning_rate` ---> `tf.keras.optimizers.Adam(learning_rate=0.001)`
  * **Fitting on more data** ---> `model.fit(X_train_full, y_train_full, epochs = 5)`
  * **Fitting for longer** --->`model.fit(X_train_full, y_train_full, epochs = 100)`

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation=None),
    tf.keras.layers.Dense(1)
])

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics=["mae"])

model.fit(tf.expand_dims(X,axis=-1), y, epochs=100)
```

```python
#Create a bigger data or feature, independent variable
X = tf.range(-100, 100, 4)
X, X.shape, X.ndim
```

```python
#Create a bigger label or dependent variable
y = X + 10
y
```

```python
plt.scatter(X,y);
```

# the three sets
* **Training set** - the model learns from this data, rwhich is typically `70 - 80%` of the total data you have available.
* **Validation set** - the model gets tuned in this data, which is typicaslly `10, 15%` of the data available
* **Test set** - the model gets evaluated on this data to test what it has learned, this set is typically `10 - 15%` of the total data available

```python
len(X), len(y)
```

```python
# Split the data into train and test sets
X_train = X[:40] # first 40 are the training samples (80% of the data)
y_train = y[:40]

X_test = X[40:] # last 10 are the testing samples (20% of the data)
y_test = y[40:]

len(X_train), len(X_test)
```

```python
#visualizing the data
plt.figure(figsize=(10,7))
plt.scatter(X_train, y_train, c="b", label = "Training data")
plt.scatter(X_test, y_test, c="g", label = "Test data")
plt.legend();
```

```python
tf.random.set_seed(42)
#create a model
model = tf.keras.Sequential(
    tf.keras.layers.Dense(1,input_shape=[1])
)

#compile a model
model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(learning_rate=0.001),
              metrics = (["mae"]))
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=100,verbose=1)
```
