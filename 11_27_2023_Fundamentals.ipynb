## Creating tensor using `tf.constant()`

### Importing library

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

print("tensorflow version:",tf.__version__)
print("numpy version:",np.__version__)
```

### Creating scalar

```python
scalar = tf.constant([7],dtype=tf.float32)
print("scalar value:",scalar.numpy())
print("scalar shape:",scalar.shape)
print("scalar datatype:", scalar.dtype)
print("scalar size:",tf.size(scalar).numpy())
print("scalar dimension",scalar.ndim)
```

### Creating vector

```python
vector = tf.constant([10,7],dtype=tf.float32)
print("vector value:",vector.numpy())
print("vector shape:",vector.shape)
print("vector datatype:",vector.dtype)
print("vector size:",tf.size(vector).numpy())
print("vector dimension:",vector.ndim)
```

## Shifting tensors using:
  1. `tf.random.shuffle()`
  2. `tf.transpose()`

```python
tf.random.set_seed(42)
x = tf.random.uniform([2,3],dtype=tf.int32,minval=0,maxval=5,seed=1)
y = tf.random.uniform([3,3],dtype=tf.int32,minval=0,maxval=10,seed=1)
x,y
```

```python
shuffled_x = tf.random.shuffle(x)
shuffled_y = tf.random.shuffle(y)
shuffled_x, shuffled_y
```

```python
transpose_x = tf.transpose(x)
transpose_y = tf.transpose(y)
transpose_x, transpose_y
```

```python
tf.reshape(x,[3,2])
```

## Tensor manipulation using:
1. `tf.add` or `tf.math.add` or `+`
2. `tf.subtract` or `tf.math.subtract` or `-`
3. `tf.multiply` or `tf.math.multiply` or `*`
4. `tf.divide` or `tf.math.divide` or `/`
5. `tf.matmul`
6. `tf.tensordot`
7. `@`

```python
tf.random.set_seed(42)
x = tf.random.uniform([3,3],dtype=tf.int32,maxval=50,minval=0,seed=1)
y = tf.random.uniform([2,3],dtype=tf.int32,maxval=100,minval=0,seed=1)
x, y
```

```python
print(tf.add(x,10))
print(tf.math.add(x,10))
print(x+10)
```

```python
print(tf.subtract(x,10))
print(tf.math.subtract(x,10))
print(x-10)
```

```python
print(tf.multiply(x,10))
print(tf.math.multiply(x,10))
print(x*10)
```

```python
print(tf.divide(x,10))
print(tf.math.divide(x,10))
print(x/10)
```

```python
print(tf.matmul(x,tf.transpose(y)))
print(tf.tensordot(x,tf.transpose(y),1))
```

## Getting the tensor's:
1. minimum using `tf.reduce_min`
2. maximum using `tf.reduce_max`
3. mean using `tf.reduce.mean`
4. sum using `tf.reduce_sum`
5. variance using `tf.math.reduce_variance`
6. std_dev using `tf.math.reduce_std`

```python
tf.random.set_seed(42)
x = tf.random.uniform([4,4],dtype=tf.int32,maxval=100,minval=50,seed=1)
x
```

```python
print("minimum value of x is:",tf.reduce_min(x))
print("Maximum value of x is:",tf.reduce_max(x))
print("mean value of x is:",tf.reduce_mean(x))
print("sum of x is:", tf.reduce_sum(x))
print("variance of x is:",tf.math.reduce_variance(tf.cast(x,tf.float32)))
print("std dev of x is:",tf.math.reduce_std(tf.cast(x,tf.float32)))
```
